```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv("archive/tmdb_5000_credits.csv")
movies_df = pd.read_csv("archive/tmdb_5000_movies.csv")
```

```python
credits.head()
```

```python
movies_df.head()
```

```python
print("credits:", credits.shape)
print("movies:", movies_df.shape)
```

```python
credits_column_renamed = credits.rename(index=str, columns={"movie_id": "id"})
movies_df_merge = movies_df.merge(credits_column_renamed, on='id')
movies_df_merge.head()
```

```python
cleaned_movies_df = movies_df_merge.drop(columns=['homepage', 'title_x', 'title_y', 'status', 'production_countries'])
cleaned_movies_df.head()
```

```python
cleaned_movies_df.info()
```

```python
v = cleaned_movies_df['vote_count']
R = cleaned_movies_df['vote_average']
C = cleaned_movies_df['vote_average'].mean()
m = cleaned_movies_df['vote_count'].quantile(0.70)
```

```python
cleaned_movies_df['weighted_average'] = ((R*v) + (C*m)) / (v + m)
cleaned_movies_df.head()
```

```python
sorted_rank_movie_df = cleaned_movies_df.sort_values('weighted_average', ascending=False)
sorted_rank_movie_df[['original_title', 'vote_count', 'vote_average', 'weighted_average', 'popularity']].head(20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

weighted_average = sorted_rank_movie_df.sort_values('weighted_average', ascending= False)
plt.figure(figsize=(12,6))
axis1 = sns.barplot(x=weighted_average['weighted_average'].head(10), y=weighted_average['original_title'].head(10))
plt.xlim(4, 10)
plt.title('Best Movies by average votes', weight='bold')
plt.xlabel('weighted Average Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('best_movies.png')
```

```python
popularity = sorted_rank_movie_df.sort_values('popularity', ascending=False)
plt.figur=(12,6)
ax=sns.barplot(x=popularity['popularity'].head(10), y=popularity['original_title'].head(10), data=popularity)

plt.title('Most Popular by votes', weight='bold')
plt.xlabel('Popularity Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('popular_movies.png')
```

```python
popularity.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
movie_scaled_df = scaler.fit_transform(cleaned_movies_df[['weighted_average', 'popularity']])
movie_normalized_df = pd.DataFrame(movie_scaled_df, columns=['weighted_average', 'popularity'])
movie_normalized_df.head()
```

```python
cleaned_movies_df[['normalized_weighted_average', 'normalized_popularity']] = movie_normalized_df
cleaned_movies_df.head()
```

```python
cleaned_movies_df['score'] = cleaned_movies_df['normalized_weighted_average']*0.5 + cleaned_movies_df['normalized_popularity']*0.5
movie_score_df = cleaned_movies_df.sort_values(['score'], ascending=False)
movie_score_df[['original_title', 'normalized_weighted_average', 'normalized_popularity', 'score']].head()
```

```python
score_df = cleaned_movies_df.sort_values('score', ascending=False)
plt.figure(figsize=(16,6))
ax = sns.barplot(x=score_df['score'].head(10), y=score_df['original_title'].head(10), data=score_df)

plt.title('Best Sore Movie', weight='bold')
plt.xlabel('Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')

plt.savefig('scored_movies.png')
```
