```python
# Minimizing the data for better analysis

selective_data = moviedata[(moviedata['avg_rating'] >= 7) &
                        (moviedata['domestic_rev'] >= 25) &
                        (moviedata['start_year'] > 2012)]

selective_data
```

```python
selective_data = selective_data.assign(genres = selective_data['genres'].str.split(',')).explode('genres')

selective_data
```

## DATA ANALYSIS

### Number of Movies in each Genre

```python
# Let us take the "Title" and "Genres" columns for the dataframe we have cleaned and assign them to "Movie_Genres".

movie_genres = selective_data[['title','genres']]
movie_genres.info()
```

```python
# Let us check the data for NAN values

movie_genres.isnull().sum()
movie_genres.head()
```

```python
# Let us now create a new dataframe that shows the number of movies per genre from our dataset.

movie_genres_count = (pd.DataFrame(movie_genres.groupby('genres', as_index = False).title.nunique().sort_values(
    by = 'title', ascending = False)))
movie_genres_count
```

```python
# Let us check the dataframe for NAN values again.

movie_genres_count.isnull().sum()

movie_genres_count.info()
```

```python
# Let us rename the column "title" to "number_of_movies" for more clarity.

movie_genres_count = movie_genres_count.rename(columns = {'title': 'number_of_movies'})
movie_genres_count
```

```python
# Let us investigate the dataframe further.

movie_genres_count.describe()
```

## VISUALIZE THE DATA

```python
# Let us plot the data for the movies in the dataframe we just created above.

plt.figure(figsize = (20, 10))
sns.barplot(x = 'genres', y = 'number_of_movies', data = movie_genres_count, )
sns.set_context('talk', rc = {'grid.linewidth': 1})
plt.xticks(rotation = 90)
plt.xlabel('Genres', fontsize = 18)
plt.ylabel('Number of Movies', fontsize = 18)
plt.title('Number of Movies in Each Genre', fontsize = 25)
```

```python
# Let us create a variable called "Ratings Per Genre" that holds the Genres and the Average rating for each genre.
# While we are at it let us also check the data for missing or NAN values.

ratings_per_genre = selective_data[['genres', 'avg_rating']]
ratings_per_genre.isnull().sum()
ratings_per_genre.info()
ratings_per_genre.describe()
```

```python
# # Now let us see how the average ratings are for each genre in our dataset.
# # First we need to create a dataframe that shows the average rating for each genre.

genre_rating = (pd.DataFrame(ratings_per_genre.groupby('genres', as_index = False).avg_rating.agg(['count', 'mean']).sort_values(
      by = 'mean', ascending = False)))

genre_rating
```

```python
# Let us keep our data clean by filtering out the genres that have less than 10 counts.

genre_rating = genre_rating[genre_rating['count'] >= 10]
genre_rating.sort_values(by = 'mean', ascending = False)

genre_rating
```

```python
# Let us visually display the results by plotting the genre rating.

plt.figure(figsize = (20, 10))
sns.lineplot(x = 'genres', y ='mean', data = genre_rating)
sns.set_context('talk', rc = {'grid.linewidth': 1})
plt.xticks(rotation = 90)
plt.xlabel('Genres', fontsize = 18)
plt.ylabel('Average Rating', fontsize = 18)
plt.title('Average Ratings Per Genre', fontsize = 25)
```

From the above data we observe "Sci-Fi", "History", and "Biography" consecutively are the top three categories based on their rating. Drama however with a rating of 7.5 leads by a huge number of counts. This requires further investigation to determine whether this is correct or not.

```python
# Let us check the mean, median and standard deviation of the genres' ratings. Since we are looking at a genre with a number of ratings, we will not be calculating the mode for now.
genre_rating.sort_values(by = 'mean')
print(f"The mean is {genre_rating['mean'].mean()}")
# print(f"The mode is {genre_rating['mean'].mode()}")
print(f"The median is {genre_rating['mean'].median()}")
print(f"The standard deviation is {genre_rating['mean'].std()}")

      
# genre_rating['mean'].median()
# genre_rating['mean'].mode()
# genre_rating['mean'].std()
```

```python
# Let us plot the movies by their genres and domestic revenue.
genre_earnings = (pd.DataFrame(selective_data.groupby('genres', as_index = False).domestic_rev.agg(['count', 'mean']).sort_values(
    by ='mean', ascending = False)))
genre_earnings 

# let us plot the earning of each genre using the genre and mean columns.

plt.figure(figsize = (20, 10))
sns.barplot(x = 'genres', y ='mean', data = genre_earnings)
sns.set_context('talk', rc = {'grid.linewidth': 1})
plt.xticks(rotation = 90)
plt.xlabel('Genres', fontsize = 18)

```

Ok from the above plot we learn that when it comes to domestic earnings(incomea), the high ranking genres are as follows:
1. SciFi
2. Animation
3. Adventure
4. Fantasy
5. Action

These five genres are shown here to be earning way more than the rest. The genres registering least domestic earnings i.e. less than 100 million are:
1. Romance
2. Crime
3. Mystery
4. Drama
5. History

This is looking interesting because according to the average rating Drama was the highest but interms of income it is one of the least earning. So the hypotheses is that a lot of the movies are categorized as Dramas. You should be careful not to be fooled by it then.


```python
# Ok now let us narrow down the genre earnings to 5 years inorder to get a more accurate understanding.
# For this is let us create a dataframe with the genre, domestic revenue and year columns.

top_earning_genres = pd.DataFrame(selective_data[['genres', 'domestic_rev', 'start_year']])
top_earning_genres = top_earning_genres[(top_earning_genres['domestic_rev'] >= 150.00) & 
                                        (top_earning_genres['start_year'] > 2012)]

top_earning_genres.head()
# Let us create a new column called "earning_bil" that holds the domestic revenue in billions.



```

```python
# Let us create a new column called "earning_bil" that holds the domestic revenue in billions.
top_earning_genres = top_earning_genres.groupby('genres', as_index = False).sum()
top_earning_genres['earning_bil'] = (top_earning_genres['domestic_rev'])/1000
top_earning_genres.sort_values('domestic_rev', ascending = False, inplace = True)
top_earning_genres.drop(columns = ['start_year'], inplace = True)

top_earning_genres.info()
top_earning_genres
```

```python
# Let us plot the results of the above dataframe.

plt.figure(figsize = (20, 10))
sns.barplot(x = 'genres', y = 'earning_bil', data = top_earning_genres)
sns.set_context('talk', rc = {'grid.linewidth': 1})
plt.xticks(rotation = 90)
plt.xlabel('Genres', fontsize = 18)
plt.ylabel('Earnings in Billions', fontsize = 18)
plt.title('Top Earning Genres 2013-2019', fontsize = 25)
```


From these findings we can clearly see that as of 2013 onwards Adventure and Action genres are high earning movie genres. While Romance and Mystery are the least earning ones. Based on this information we can deduce that currently Adventure and Action genres are trendy and highly demanded genres in the movie business.
