# APIs

The term **Application Program Interfaces**, or APIs, is exceedingly general. It applies to any interaction between applications or between an application and a user. One might speak of the "matplotlib API" to describe proper plotting syntax but also of APIs that govern the interaction between various steps of a data pipeline: between a database server and a remote client, between a local machine and a cloud storage bucket, etc.

APIs are commonly used to retrieve data from remote websites. Sites like Reddit, Twitter, and Facebook all offer certain data through their APIs.

## Overview of APIs

An API communicates with another application:

* Send request (with some info/data)
* Get response
    + data
    + service


It's always a software-to-software interaction

### Parts on an API

* **Access Permissions**
    + User allowed to ask?
* **API Call/Request**
    + Code used to make API call to implement complicated tasks/features
    + *Methods*: what questions can we ask?
    + *Parameters*: more info to be sent
* **Repsonse**
    + Result of request

## Status Codes

The request we make may not always be successful. The best way is to check the status code which gets returned with the response: `response.status_code`

````python
comments_response.status_code
````

[Status Code Info](https://developer.mozilla.org/en-US/docs/Web/HTTP/Status) <br/>
[Wikipedia on Status Codes](https://en.wikipedia.org/wiki/List_of_HTTP_status_codes)

So this is a good check to see if our request was successful. Depending on the status of the web server, the access rights of the clients and availibility of requested information. A web server may return a number of status codes within the response. Wikipedia has exhaustive details on all these codes.

### Common status codes

* 200 — everything went okay, and the result has been returned (if any)
* 301 — the server is redirecting you to a different endpoint. This can happen when a company switches domain names, or an endpoint name is changed.
* 401 — the server thinks you’re not authenticated. This happens when you don’t send the right credentials to access an API.
* 400 — the server thinks you made a bad request. This can happen when you don’t send along the right data, among other things.
* 403 — the resource you’re trying to access is forbidden — you don’t have the right permissions to see it.
* 404 — the resource you tried to access wasn’t found on the server.
* 500 — the server encountered an unexpected condition that prevented it from fulfilling the request (internal server error).
* 503 — the server is currently unable to handle the request due to a temporary overload or scheduled maintenance, which will likely be alleviated after some delay.

## Query Parameters

If you look at the [documentation](https://developer.themoviedb.org/reference/movie-popular-list), we see that the TMDB Popular Movies endpoint can take 3 parameters.

We can add parameters by adding an optional keyword argument (params) to our request. In this case, we can use two parameters:

* language — The languagew e want the response to be in.
* page — The number of pages we want returned (pagination).

We can make a dictionary with these parameters, and then pass them into the `requests.get()` method. We’ll make a request to get results in the English language and receive 1 page of results, and see what response we get.

````python
# ecac84ea73741be467e2507e4bc2d2f5
````

````python
# Our code here
popular_movies = "https://api.themoviedb.org/3/movie/popular?api_key="

response = requests.get(popular_movies ,
            params={'language': 'en-US', 'page': 1})

# Print the content of the response (the data the server returned)

print(response.json)

# This gets the same data as the following command:
# curl --request GET \ --url 'https://api.themoviedb.org/3/movie/popular?api_key=API_KEY&language=en-US&page=1' \ --header 'accept: application/json'api_key=*.json
````

## Making our Request

[TMDB API Documentation](https://developer.themoviedb.org/docs)

Let's look at an example request and dissect it into its consituent parts:

```python
popular_movies = f"https://api.themoviedb.org/3/movie/popular?api_key={creds['key']}"

headers = {
    "accept": "application/json"
}

url_params = {
    'language': 'en-US',
    'page': 1
}

response = requests.get(popular_movies, headers=headers, params=url_params)


# Print the content of the response (the data the server returned)

print(response.content)
```

## Breaking Down the Request

As you can see, there are three main parts to our request.  
  
They are:
* The URL
* The header
* The parameters
  
The URL is found in the documentation (`https://developer.themoviedb.org/reference`) and we are using the Popular endpoint (`/popular`) from the main address (`https://api.themoviedb.org/3/movie/`)

The `API KEY` is required by the TMDB API for authorization.

The header structure is used in the context of making HTTP requests, where they are utilized to provide additional information to the server about the client's request. The "accept" header specifically indicates the media type that is acceptable for the response.

In this case, the value `"application/json"` suggests that the client is indicating its preference for a JSON response. This information is essential for the server to understand the type of content that the client expects in the response. By setting the `"accept"` header to `"application/json,"` the client is indicating that it can handle and process JSON data.

The parameters contain information/options we pass into the query to get the data we want. Valid parameters by which to structure your queries, are described in the [TMDB API Documentation](https://developer.themoviedb.org/reference). We make a `url_params` dictionary to pass into our `.get()` method, which then adds the query parameters to the URL.
